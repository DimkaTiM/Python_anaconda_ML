```python
def all_comment_bank(link, n=10): 
    # –ù–∞ –≤—Ö–æ–¥ —Å—Å—ã–ª–∫–∞ –±–∞–Ω–∫–∞
    # –ù–∞ –≤—ã—Ö–æ–¥ –º–∞—Å—Å–∏–≤ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤ c 10 —Å—Ç—Ä–∞–Ω–∏—Ü 
    rec = requests.get(link)
    soup = BeautifulSoup(rec.text, 'html.parser')
    bank_name = soup.find("div", {"class": "header-h2 display-inline margin-right-x-small"}).get_text().strip()
    text_comments = ""
    for i in range(2, n):
        l = link + "?page=" + str(i)
        rec = requests.get(link)
        soup = BeautifulSoup(rec.text, 'html.parser')
        a = soup.find("div", {"class": "padding-default bg-gray mobile-full-width"}).find_all("div", {"class":"responses__item__message markup-inside-small markup-inside-small--bullet"})
        for i in a:
            text_comments += i.get_text().strip()
    return text_comments.replace("\xa0", "")
```
