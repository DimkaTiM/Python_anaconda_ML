```python
from nltk.corpus import stopwords
from nltk.corpus import RegexpTokenizer
import requests
from bs4 import BeautifulSoup
import nltk
```

```python
def get_soup(link):
    """ Получаем soup с ссылки
    :param link: ссылка
    :return: soup
    """
    con = requests.get(link).content
    soup = BeautifulSoup(con, "lxml")
    return soup
```

```python
def get_page_links(obj):
    """Все ссылки со страницы
    :param obj: soup
    :return: ссылки со страницы
    """
    links = []
    for lin in obj:
        text = "http://firrma.ru/data/"
        links.append(text + lin['href'])
    return links
```

```python
link  = "http://firrma.ru/data/"
soup = get_soup(link)
obj2 = soup.find('div', {'class': ['news-list']}).find_all("div", {'class' : 'img'})
title = []
for ent in obj2:
    title.append(ent.img.get("alt"))
```
